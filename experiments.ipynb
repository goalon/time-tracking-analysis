```python
from datetime import datetime
import jsonlines
import matplotlib.pyplot as plt
```

```python
DATA_FILE_PATH = './data/2022/10/12.jsonl'
```

```python
with jsonlines.open(DATA_FILE_PATH) as reader:
    raw_data = list(reader)
```

```python
raw_data
```

```python
additions = [v for d in raw_data for v in [0, d['additions']]]
deletions = [-v for d in raw_data for v in [0, d['deletions']]]
dates = [datetime.fromisoformat(v).timestamp() for d in raw_data for v in [d['start'], d['end']]]
```

```python
plt.figure(figsize=(35, 18))
plt.step(dates, additions)
plt.step(dates, deletions)
```

```python
def hours_to_arr(hours):
  hours_arr = []
  for i in range(0, 24):
      if f'{i}' in hours:
        hours_arr.append(hours[f'{i}'])
      else:
        hours_arr.append(0)
  return hours_arr
```

```python
plt.figure(figsize=(35, 18))
plt.bar(list(range(0, 24)), hours_to_arr(hours))
```
